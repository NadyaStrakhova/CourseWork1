
Задача: Проводится эксперимент. Частица совершает случайное блуждание, где приращение имеет нормальное распределение:
$S_{0} = 0, S_{n} = \sum_{i = 1}^{n} X_{i}$, $X_{i} \sim N(0, \sigma ^2)$,  и мы знаем $\sigma^2$.

В какой-то момент мы фиксируем ее попадание в луч и начинаем замерять время $t$ ее пребывания там. Необходимо оценить ширину луча $d$ при известных $t$ и $\sigma$.


До попадания молекулы в луч, она совершает случайное блуждание. Введем обозначения:
$$S_{0} = 0, S_{n} = \sum_{i = 1}^{n} X_{i}$$ где $X_{i}$ имеет нормальное распределение $N(0, \sigma ^2)$.


Пусть $a$ -  координата начала луча, которую мы задаем сами. Введем следующие обозначения:
$$N(a) = \max\{n \geq 1: S_{n} \leq a\}$$
Таким образом, координата, с которой наша молекула начинает блуждать в луче, будет обозначаться 
    $ X_{0} = S_{N(a) + 1} - a $ - наш эксцесс.


Введем понятия первого лестничного индекса и первой лестничной выстоы:
$$ \eta_{+} = \max\{ n \geq 1: S_n \geq 0\}, \chi_{+} = S_{\eta_{+}}$$
 
 
 
Понятно, что мы можем представить $S_{N(a) + 1} = \chi_{1} + \dots + \chi_{k}$, где $\chi_{i}$ 
   распределены как  $\chi_{+}$. 
  
  
Заметим, что теперь мы можем воспольоваться результатами теории восстановления, потому что $\chi_{i} \geq 0$.
Для эксцесса и дефекта в теории восстановления у нас есть следующая теорема:
    

Пусть $X_{i}$ имеет непрерыное распределение, тогда, обозначив за $\Delta_{1} = a - S_{N(a)}$ - дефект, а за
        $\Delta_{2} = S_{N(a) + 1} - a$ - эксцесс. Toгда имеем: $\forall u, v \geq 0$
$$ \lim_{a\rightarrow \infty}P\left(\Delta_{1} > u, \Delta_{2} > v \right) = \dfrac{1}{EX}\int_{u + v}^{\infty} P(X > x) dx$$
В стастье В. И. Лотова "On some boundary crossing problems for Gaussian Random Walks" 
для нашего случая подсчитаны следующие моменты случайной величины $\chi_{+}$:
$$E\chi_{+} = \dfrac{\sigma}{\sqrt{2}}$$
$$E\chi_{+}^{2} = \dfrac{K\sigma^2}{\sqrt{\pi}}$$
$$E\chi_{+}^{3} = \dfrac{3\sigma^{3}}{4\sqrt{2}} + \dfrac{3K^2\sigma^3}{2\sqrt{2}\pi}$$ где
$K = 1.460\dots$, взята из равенства $\sum_{m=1}^ {n}\dfrac{1}{\sqrt{m}} = 2\sqrt{n} - K + O\Big(\dfrac{1}{\sqrt{n}}\Big)$
    

Оценим, к чему стремятся $E(S_{N(a) + 1} - a)$ и $E(S_{N(a) + 1} - a)^2$ при $a \rightarrow \infty$.

    
 $$\lim_{a\rightarrow\infty} E(S_{N(a) + 1} - a) = \int_{0}^{\infty}\lim_{a\rightarrow\infty}P(\Delta_{2} > u) du =
    \dfrac{1}{E\chi_{+}}\int_{0}^{\infty}\int_{u}^{\infty}P(\chi_{+} > x) dx du = \dfrac{1}{E\chi_{+}}\int_{0}^{\infty}
    \int_{0}^{x}P(\chi_{+}> x) du dx = $$ 
    $$= \dfrac{1}{E\chi_{+}}\int_{0}^{\infty}xP(\chi_{+} > x)dx$$
   Рассмотрим
    $$E\chi_{+}^2 = \int_{0}^{\infty}x^2 dP(\chi_{+} \leq x) = - \int_{0}^{\infty} x^2 dP(\chi_{+} > x) = 2\int_{0}^{\infty}xP(\chi_{+} > x)dx$$
   Следовательно,
    $$\lim_{a\rightarrow\infty} E(S_{N(a) + 1} - a) = \dfrac{E\chi_{+}^2}{2E\chi_{+}}$$
   Теперь рассмотрим:
    $$\lim_{a\rightarrow\infty}E(S_{N(a) + 1} - a)^2 = 
    \lim_{a\rightarrow\infty} 2\int_{0}^{\infty} uP(\Delta_{2} > u) du = 
    \dfrac{2}{E\chi_{+}}\int_{0}^{\infty}u\int_{u}^{\infty}P(\chi_{+} > x)dxdu = 
    \dfrac{2}{E\chi_{+}}\int_{0}^{\infty}\int_{0}^{x}uP(\chi_{+} > x) du dx = $$ $$ = 
    \dfrac{1}{E\chi_{+}}\int_{0}^{\infty} x^2P(\chi_{+} > x) dx$$
   Для дальнейшего вычисления посмотрим на равенство:
    $$E\chi_{+}^3 = \int_{0}^{\infty} x^3 dP(\chi_{+}\leq x) = -\int_{0}^{\infty} x^3 dP(\chi_{+} > x) = 3\int_{0}^{\infty}  x^2 P(\chi_{+} > x)dx$$
   Отсюда получаем:
    $$\lim_{a\rightarrow\infty}E(S_{N(a) + 1} - a)^2 = \dfrac{E\chi_{+}^3}{3E\chi_{+}}$$

Теперь рассмотрим вторую часть нашего эксперимента, где молекула попадает в луч. 
Сделаем сдвиг на $a$. Теперь наш луч имеет границы $0$ и $d$.


Молекула начинает свое движение из точки $X_{0} = S_{N(a) + 1} - a$. 
Опишем ее движение так: $\Gamma(t) = W_{t} + X_{0}$, где $W_{t}$ - броуновское движение.
Так как, $W_{t}$ мартингал, то $\Gamma(t)$ тоже будет мартингалом, так как $X_{0}$ случайная константа. 


Введем $\tau$ - cлучайную величину, равную первому моменту выхода из луча. Это будет марковский момент.
По теореме об остановке мартингала мы получаем $E\Gamma(\tau) = E\Gamma(0) = EX_{0}$. 
С другой стороны $E\Gamma(\tau) = dP_{1}$, где $P_{1}$ - вероятность молекулы пересечь верхнюю границу луча.


Рассмотрим еще один процесс: $\Gamma(t)^2 - \sigma^2 t = W_{t}^2 - ^2\sigma t + 2X_{0}W_{t} + X_{0}^2$ - это тоже будет мартингалом, так как $W_{t}^2 - \sigma t$ мартингал.
По той же самой теореме получим: $E(\Gamma(\tau)^2 - \sigma^2 \tau) = E(\Gamma(0)^2) = EX_{0}^2$. 
Тут же по линейности у нас выходит $E(\Gamma(\tau)^2 - \sigma^2\tau) = E\Gamma(\tau)^2 - \sigma^2 E\tau = d^2P_{1} - \sigma^2 E\tau$.
Теперь у нас есть следующая система:
    $$ dP_{1} = EX_{0} $$
    $$ d^2P_{1} = EX_{0}^2 + \sigma^2 E\tau $$
    $$ E\tau = \overline{T}$$
    
Решая эту систему методом моментов, мы получаем оценку $ \hat d = \dfrac{EX_{0}^2 + \sigma^2 \overline{T}}{EX_{0}} $ для равенства $d = \dfrac{\sigma^2 E\tau + EX_{0}^2}{EX_{0}}$
А отсюда, при $a\rightarrow\infty$ имеем
$\hat d \rightarrow \dfrac{2(E\chi_{+}^3 + \sigma^2 3E\chi_{+}\overline{T})}{3E\chi_{+}^2}$


Теперь мы хотим доказать, что эта оценка асимптотически нормальная,
и посторить доверительный интервал для ширины луча $d$.
Для оценки методом моментов у нас есть следующая теорема:
   
   
   Если теоретический момент $EX^{2k}$ конечен, то при $n\rightarrow\infty$ оценка
    $X_{n}^{k} = \dfrac{1}{n}\sum_{i=1}^{n} X_{i}^{k}$ для параметра $EX^k$ будет асимптотически нормальная
    $N\Big( EX^k, \dfrac{EX^{2k} - (EX^k)^2}{n}\Big)$.
    

Если теперь предположить, что $E\tau^2 < \infty$, то оценка $\overline{T}$ будет асимтотически нормальна 
$N\Big( E\tau, \dfrac{E\tau^2 - (E\tau)^2}{n}\Big)$.
 Отсюда мы получаем $ d = \dfrac{EX_{0}^2 + \sigma \overline{T}}{EX_{0}} $ асимтотически нормальна
    $N\Big(\dfrac{\sigma^2 E\tau + EX_{0}^2}{EX_{0}}, \dfrac{\sigma^4 E\tau^2 - \sigma^4 (E\tau)^2}{n(EX_{0})^2} \Big)$.
    Построим доверительный интервал : $(T_{1}(X), T_{2}(X))$ такой, что $P(T_{1}(X) \leq d \leq T_{2}(X)) =  \gamma$.
Для $E\tau$ возьмем оценку $\overline{T}$, а для $D\tau^2$ возьмем $S = \dfrac{1}{n}\sum_{i=1}^{n}(T_{i} - \overline{T})^2$. 
Из-за асимптотической нормальности нашей оценки мы можем взять для заданного $\gamma$ следующий доверительный интервал:
    
$\Big(\dfrac{\sigma^2 \overline{T} + EX_{0}^2}{EX_{0}} - \dfrac{c_{\gamma}\sigma^2\sqrt{S}}{\sqrt{n}EX_{0}}, \dfrac{\sigma^2 \overline{T} + EX_{0}^2}{EX_{0}} + \dfrac{c_{\gamma}\sigma^2 \sqrt{S}}{\sqrt{n}EX_{0}}\Big)$
 где величина $c_{\gamma} = \Phi^{-1}((1 + \gamma)/2)$ 
$ \Big( \Phi(x) = \dfrac{1}{\sqrt{2\pi}\sigma}\int_{-\infty}^{x} \exp^{-\dfrac{x^2}{2}}dx \Big) $ взята из таблицы.


```python
import numpy as np
from scipy.stats import norm
from math import sqrt
import matplotlib.pyplot as plt
```

```python
def mean(arr):
    l = len(arr)
    return sum(arr) / l
```


Функция $mean$ получает на вход массив значений $X_{1}, X_{2}, \dots$ и считает $\overline{X}$. Это несмещенная и состоятельная оценка для $EX$

```python
def variable(arr):
    l = len(arr)
    summa_2 = 0
    for i in arr:
        summa_2 += i * i
    summa_1 = mean(arr)
    return (summa_2 / l) - (summa_1 ** 2)
```


Функция $variable$ получает на вход массив $X_{1}, X_{2}, \dots$ и считает $S = \dfrac{1}{n} \sum_{i = 1}^{n} (X_{i} - \overline{X}) = \dfrac{1}{n}\sum_{i = 1}^{n} X_{i}^2 - \overline{X}^2$ -  Состоятельная оценка для $DX$, так как $S_{0} = \dfrac{n}{n-1}S$ - несмещенная и состоятельная оценка.

```python
def second_moment(arr):
    l = len(arr)
    summa_2 = 0
    for i in arr:
        summa_2 += i * i
    return summa_2 / l
```

 Функция $second \ moment$ получает на вход массив данных $X_{1}, X_{2} \dots $ и вычисляет $X_{n}^{2} = \dfrac{1}{n}\sum_{i=1}^{n} X_{i}^2$

```python
def excess_first_experiment(a, sigma):
    rand_walk_cur = 0
    while rand_walk_cur < a:
        rand_walk_cur += norm.rvs(scale=sigma)
        if rand_walk_cur < -a:
            rand_walk_cur = 0
    return rand_walk_cur - a
```

Функция $excess\ first\ experiment$ получает на вход $a$ - координата, где начинается наш луч, и $\sigma$ - дисперсия распределения. 
Далее генерируется случайное блуждание, после чего возвращается значение $S_{N(a) + 1} - a$.

```python
def time_second_experiment(begin, dt, sigma, d):
    time_cur = 0
    pos_cur = begin
    while 0 < pos_cur < d:
      #  pos_cur = pos_cur + norm.rvs(scale=sigma**2*dt)
       # pos_cur = pos_cur + norm.rvs(scale=dt)
        pos_cur = pos_cur + norm.rvs(scale=sigma*sqrt(dt))
        time_cur += dt
    if pos_cur > d:
        return time_cur, 1
    else:
        return time_cur, 0
```

Функция $time\ second \ expriment $ получает на вход: $begin$ - координату начального положуния,
$dt$ - приращение времени $\sigma$ - дисперсия (нужна ли она тут??) $d$ - ширина луча. 
Далее функция генерирует траекторию $\Gamma(t) = X_{0} + W_{t}$ и возвращает $\tau$ - момент времени, когда молекула выходит из нашего луча

```python
def experiment(d, a, sigma, n, dt, number):
    T = []
    begin_pos = []
    count_d = 0
    for i in range(n):
        x0 = excess_first_experiment(a, sigma)
        begin_pos.append(x0)
        t, c = time_second_experiment(x0, dt, sigma, d)
        count_d += c
        T.append(t)
    mean_begin = mean(begin_pos)
    mean_T = mean(T)
    print("{} experiment: average excess {:.5f}, average time {:.5f}, number of crossing d {}".format(number, mean_begin, mean_T, count_d))
    return begin_pos, T
```

Функция $experiment$ получает на вход $d$ - наш заданный диаметр луча и то, что мы пытаемся оценить,
$a$ - координата, где начинается наш луч, $sigma$ - заданная вариация, $n$ - количество экспериментов
$dt$ - это для броуновского движения, $number$ - номер эксперимента, программа генерирует 2 эксперимента и создает 2 массива:
$begin\ pos$ - массив эксцессов при первой части, $T$ - массив из наблюдаемого времени, сколько наша молекула пробывала в луче

```python
def our_experiment(d, a, sigma, num, n):
    K = 1.460
    pi = 3.14
    '''determine accuracy of our interval'''
    '''\gamma = 0.9, c_\gamma = 1.645'''
    C = 1.645
    mean_chi = sigma / sqrt(2)
    mean_chi2 = K * (sigma ** 2) / sqrt(pi)
    mean_chi3 = ((3 * (sigma ** 3)) / (4 * sqrt(2))) + ((3 * K**2 * sigma**3) / (2 * sqrt(2) * pi))
    mean_excess2 = mean_chi3 / (3 * mean_chi)
    mean_excess = mean_chi2 / (2 * mean_chi)
    print('our {} experiment: our initial values: a = {:.5f}, d = {:.5f}, sigma = {:.5f}, n = {}'.format(num, a, d, sigma, n))
    begin, T = experiment(d=d, a=a, sigma=sigma, dt=0.01, number = num, n = n)
    print('expected mean excess {:.5f}, average excess{:.5f}, difference {:.5f}'.format(mean_excess, mean(begin), abs(mean_excess - mean(begin))))
    print('expected mean^2 excess {:.5f} average mean^2 excess {:.5f}, difference {:.5f}'.format(mean_excess2, second_moment(begin), abs(mean_excess2 - second_moment(begin))))
    centre = (sigma**2 * mean(T) + mean_excess2) / mean_excess
    print('our assessment of d = {:.5f} is {:.5f}'.format(d, centre))
    left_board = centre - (sigma**2 * sqrt((variable(T)) * C) / (sqrt(len(T)) * mean_excess))
    right_board = centre + (sigma**2 * sqrt((variable(T)) * C) / (sqrt(len(T)) * mean_excess))
    print('our known d is {:.5f} our interval is ({:.5f}, {:.5f})'.format(d, left_board, right_board))
    print('$\sigma = {}, d = {}, n = {} $ & ${:.5f}$ & $({:.5f}, {:.5f})$ '.format(sigma, d, n, centre, left_board, right_board))
    print('$\sigma = {}, a = {}$ & ${:.5f}$ & ${:.5f}$ & ${:.5f}$ & ${:.5f}$ \\'.format(sigma, a, mean_excess, mean_excess2,  mean(begin), second_moment(begin)))
```

Наша функция $our\ experiment$ проводит $num$ эксперимент с начальными данными: $d$- заданная ширина луча, то что мы пытаемся оценить
    $a$ - координата начала луча, $sigma$ - заданная дисперсия $n$ - количество экспериментов. 
    Вычисляем $E\chi_{+}$, $E\chi_{+}^2$, $E\chi_{+}^3$, далее вычисляем $\lim_{a\rightarrow\infty}E(S_{N(a) + 1} - a)$ 
    мат ожидание эксцесса
    $\lim_{a\rightarrow\infty}E(S_{N(a) + 1} - a)^2$ мат ожидание эксцесса в квадрате (то что мы доказывали). Потом вычисляем нашу оценку
    $\dfrac{\overline{T} + EX_{0}^2}{EX_{0}}$, И смотрим на доверительный интервал при $\gamma = 0.9, c_{\gamma} = 1.645$ 
    $\Big(\dfrac{\overline{T} + EX_{0}^2}{EX_{0}} - \dfrac{c_{\gamma}\sqrt{S}}{\sqrt{n}EX_{0}},
    \dfrac{\overline{T} + EX_{0}^2}{EX_{0}} + \dfrac{c_{\gamma}\sqrt{S}}{\sqrt{n}EX_{0}}\Big)$

```python
def for_drawing(a, d, sigma):
    fig = plt.figure()
    plt.xlabel('time n')
    plt.ylabel('position')
    begin_pos = 0
    n = 0
    while begin_pos < a:
        cur = begin_pos + norm.rvs(scale=sigma ** 2)
        plt.plot([n, n + 1],[begin_pos, cur], color='red')
        n += 1
        begin_pos = cur
        if begin_pos < -a:
            n = 0
            begin_pos = 0
    plt.plot([0, n + 1], [a, a])
    plt.plot([0, n + 1], [a + d, a + d])
    plt.show(fig)
    begin_pos -= a
    cur -= a
    fig1 = plt.figure()
    plt.xlabel('time t')
    plt.ylabel('position')
    dt = 0.01
    T = 0
    while 0 < begin_pos < d:
        cur = begin_pos + norm.rvs(scale=sigma*sqrt(dt))
        plt.plot([T, T + dt],[begin_pos, cur], color='red')
        begin_pos = cur
        T += dt
    plt.plot([0, T], [d, d])
    plt.plot([0, T], [0, 0])
    plt.show(fig1)
```
